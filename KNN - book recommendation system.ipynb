```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors
from scipy.sparse import csr_matrix
from fuzzywuzzy import process
```

```python
books = pd.read_csv('books.csv', usecols = ['book_id','title'], dtype = {'id':'int32','title':'str'})
books
```

```python
books['title'] = books['title'].replace(r"[\"\'\|\?\=\.\@\#\*\,]", '')
```

```python
ratings = pd.read_csv('ratings.csv', usecols = ['book_id','user_id','rating'], dtype = {'book_id':'int32','user_id':'int32','rating':'int32'})
ratings
```

```python
bookUsers = ratings.pivot_table(index = 'book_id', columns = 'user_id', values = 'rating').fillna(0)
bookUsers
```

```python
bookUsersMatrix = csr_matrix(bookUsers.values)
print(bookUsersMatrix)
```

```python
knnModel = NearestNeighbors(metric = 'cosine', algorithm = 'brute', n_neighbors = 20)
```

```python
knnModel.fit(bookUsersMatrix)
```

```python
def findRecommendations(book, data, model, recommendations):
    model.fit(data)
    modifyId = process.extractOne(book,books['title'])[2]
    print('BOOK : ', books['title'][modifyId], 'Index', modifyId)
    distances, indices = model.kneighbors(data[modifyId],n_neighbors = recommendations)
    for i in indices:
        print(books['title'][i].where(i!=modifyId))
```

```python
findRecommendations('little women',bookUsersMatrix,knnModel,20)
```
